```python
#Create class Guest
class Guest:
    def __init__(self, guest_id, name, email, password, phone, address, account_status):
        # Initialize guest attributes
        self.guest_id = guest_id # an id that uniquely identifies the Guest.
        self.name = name # Guest Name
        self.email = email # Guest Email
        self.password = password # Guest acc password
        self.phone = phone # Guest Number
        self.address = address # Guest Address
        self.account_status = account_status # Guest account status

    # Getter and setter methods for guest attributes
    def get_guest_id(self):
        return self.guest_id

    def set_guest_id(self, guest_id):
        self.guest_id = guest_id

    def get_name(self):
        return self.name

    def set_name(self, name):
        self.name = name

    def get_email(self):
        return self.email

    def set_email(self, email):
        self.email = email

    def get_password(self):
        return self.password

    def set_password(self, password):
        self.password = password

    def get_phone(self):
        return self.phone

    def set_phone(self, phone):
        self.phone = phone

    def get_address(self):
        return self.address

    def set_address(self, address):
        self.address = address

    def get_account_status(self):
        return self.account_status

    def set_account_status(self, status):
        self.account_status = status

#New Class, Payment class
class Payment:
    def __init__(self, payment_id, booking_id, amount, payment_method, status, transaction_date):
        # Initialize payment attributes
        self.payment_id = payment_id # a payment id that uniquely identifies.
        self.booking_id = booking_id # Booking id for payment
        self.amount = amount # paid amount
        self.payment_method = payment_method # Method of payment
        self.status = status
        self.transaction_date = transaction_date # Date of transaction

    # Getter and setter methods for payment attributes
    def get_payment_id(self):
        return self.payment_id

    def set_payment_id(self, payment_id):
        self.payment_id = payment_id

    def get_booking_id(self):
        return self.booking_id

    def set_booking_id(self, booking_id):
        self.booking_id = booking_id

    def get_amount(self):
        return self.amount

    def set_amount(self, amount):
        self.amount = amount

    def get_payment_method(self):
        return self.payment_method

    def set_payment_method(self, method):
        self.payment_method = method

    def get_status(self):
        return self.status

    def set_status(self, status):
        self.status = status

    def get_transaction_date(self):
        return self.transaction_date

    def set_transaction_date(self, transaction_date):
        self.transaction_date = transaction_date

# New Class for Booking
class Booking:
    def __init__(self, booking_id, guest_id, room_type, check_in_date, check_out_date, number_of_rooms, total_cost, discount):
        # Initialize booking attributes
        self.booking_id = booking_id #an id that uniquely identifies the booking.
        self.guest_id = guest_id # guest id for booking
        self.room_type = room_type# Type of room booked (deluxe suit)
        self.check_in_date = check_in_date # Date check in
        self.check_out_date = check_out_date # Date check out
        self.number_of_rooms = number_of_rooms #num of room (1)
        self.total_cost = total_cost # total cost 1000
        self.discount = discount #discount 15% off 850

    # Getter and setter methods for booking attributes
    def get_booking_id(self):
        return self.booking_id

    def set_booking_id(self, booking_id):
        self.booking_id = booking_id

    def get_guest_id(self):
        return self.guest_id

    def set_guest_id(self, guest_id):
        self.guest_id = guest_id

    def get_room_type(self):
        return self.room_type

    def set_room_type(self, room_type):
        self.room_type = room_type

    def get_check_in_date(self):
        return self.check_in_date

    def set_check_in_date(self, date):
        self.check_in_date = date

    def get_check_out_date(self):
        return self.check_out_date

    def set_check_out_date(self, date):
        self.check_out_date = date

    def get_number_of_rooms(self):
        return self.number_of_rooms

    def set_number_of_rooms(self, num):
        self.number_of_rooms = num

    def get_total_cost(self):
        return self.total_cost

    def set_total_cost(self, cost):
        self.total_cost = cost

    def get_discount(self):
        return self.discount

    def set_discount(self, discount):
        self.discount = discount

    # Calculate the total cost after applying the discount
    def calculate_discounted_total(self):
        return self.total_cost * (1 - (self.discount / 100))

#New Class, Receipt
class Receipt:
    def __init__(self, receipt_id, booking_id, guest_id, amount_paid, receipt_date, payment_status):
        # Initialize receipt attributes
        self.receipt_id = receipt_id # an id that uniquely identifies the reciept.
        self.booking_id = booking_id # booking id for reciept
        self.guest_id = guest_id # guest id for reciept
        self.amount_paid = amount_paid # paid amount for the room
        self.receipt_date = receipt_date # date of booking
        self.payment_status = payment_status # payment status

    # Getter and setter methods for receipt attributes
    def get_receipt_id(self):
        return self.receipt_id

    def set_receipt_id(self, receipt_id):
        self.receipt_id = receipt_id

    def get_booking_id(self):
        return self.booking_id

    def set_booking_id(self, booking_id):
        self.booking_id = booking_id

    def get_guest_id(self):
        return self.guest_id

    def set_guest_id(self, guest_id):
        self.guest_id = guest_id

    def get_amount_paid(self):
        return self.amount_paid

    def set_amount_paid(self, amount):
        self.amount_paid = amount

    def get_receipt_date(self):
        return self.receipt_date

    def set_receipt_date(self, date):
        self.receipt_date = date

    def get_payment_status(self):
        return self.payment_status

    def set_payment_status(self, status):
        self.payment_status = status

#New Class, TransferFunds
class TransferFunds:
    def __init__(self, transfer_id, amount, payment_method, status, transaction_date):
        # Initialize fund transfer attributes
        self.transfer_id = transfer_id # an id that uniquely identifies the transferedfunds.
        self.amount = amount # amount tranferred 
        self.payment_method = payment_method # method of transfer
        self.status = status # status of transfer
        self.transaction_date = transaction_date #date of transfer

    # Getter and setter methods for transfer attributes
    def get_transfer_id(self):
        return self.transfer_id

    def set_transfer_id(self, transfer_id):
        self.transfer_id = transfer_id

    def get_amount(self):
        return self.amount

    def set_amount(self, amount):
        self.amount = amount

    def get_payment_method(self):
        return self.payment_method

    def set_payment_method(self, method):
        self.payment_method = method

    def get_status(self):
        return self.status

    def set_status(self, status):
        self.status = status

    def get_transaction_date(self):
        return self.transaction_date

    def set_transaction_date(self, transaction_date):
        self.transaction_date = transaction_date


# Display the confirmation receipt
confirmation_receipt = f"""
-------------------------------------------
                  St. Regis Hotel
                Booking Confirmation
-------------------------------------------
Guest Name: {ayesha.get_name()}
Guest Email: {ayesha.get_email()}
Phone Number: {ayesha.get_phone()}
Address: {ayesha.get_address()}

Booking ID: {booking.get_booking_id()}
Room Type: {booking.get_room_type()}
Check-in Date: {booking.get_check_in_date()}
Check-out Date: {booking.get_check_out_date()}
Number of Rooms: {booking.get_number_of_rooms()}
Original Total Cost: {booking.get_total_cost()} AED
Discount: {booking.get_discount()}%
Discounted Total: {discounted_total:.2f} AED

Payment Status: {receipt.get_payment_status()}
Receipt Date: {receipt.get_receipt_date()}

Thank you for choosing St. Regis Hotel!
We look forward to welcoming you!
-------------------------------------------
"""

# Display the confirmation receipt
print(confirmation_receipt)
```
